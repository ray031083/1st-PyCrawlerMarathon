## 作業目標

根據範例 ，完成以下問題：

* ① 取出今天所有的發文
* ② 如果想要依照類別分類，怎麼儲存會比較好？
* ③ 哪一個類別的文章最多




```python
import requests
import re
import pandas as pd
from bs4 import BeautifulSoup
import datetime

url = 'http://www.ettoday.net/news/news-list.html'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")
```

### ① 取出今天所有的發文

```python
today = datetime.date.today().strftime("%Y/%m/%d")
pattern = '[0-9]{4}(?:/[0-9]{2}){2}'

t = []

for d in soup.find(class_="part_list_2").find_all('h3'):
    if re.match(pattern, d.find(class_="date").text)[0] == today:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
        t.append([d.find(class_="date").text, d.find('em').text, d.find_all('a')[-1].text])
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
columns = ['Date','Category','title']

df = pd.DataFrame(t, columns = columns)
df
```

```python
group = df.groupby('Category')
group.groups
```

### ③ 哪一個類別的文章最多

```python
df_1 = pd.DataFrame(df.groupby('Category').size(), columns=['Count'])
df_1['Rank'] = df_1.rank(method='min', ascending=False)
df_1.sort_values(by=['Rank'])
```
